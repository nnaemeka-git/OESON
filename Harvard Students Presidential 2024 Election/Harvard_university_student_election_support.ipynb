```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import pandas as pd

# Load the dataset
file_path = '/content/drive/MyDrive/OESON Internship/Linkedin/use election.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
data.head()

# Load the USA flag image
flag_image = mpimg.imread('/content/drive/MyDrive/OESON Internship/Linkedin/flag-of-the-united-states.png')
```

```python
# Create a new column combining Sex and Race for better visualization
data['Sex_Race'] = data['Sex'] + ' - ' + data['Race']
```

```python
# Count the number of supporters for each candidate by the combined Sex_Race column
candidate_sex_race_grouped = data.groupby(['Candidate', 'Sex_Race']).size().unstack().fillna(0)
```

```python
# Create the figure and axis
fig, ax = plt.subplots(figsize=(8,4))

# Plot the grouped bar chart
candidate_sex_race_grouped.plot(kind='bar', ax=ax, width=0.8)
#plt.title('Demographic Breakdown of Candidate Support: A Visualization of Combined Sex and Race Data')
plt.xlabel('')
plt.ylabel('Number of Supporters')
plt.xticks(rotation=0)
#plt.legend(title='Sex and Race', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.legend(title='Sex and Race', bbox_to_anchor=(1,1.3), loc='upper left')

# Add labels on the bars
for container in ax.containers:
    ax.bar_label(container)

# Set the background image
fig.figimage(flag_image, xo=0, yo=0, alpha=0.2, zorder=-1)
# Add footer text
fig.text(0.5, 0.01, 'Graphics By: NNAEMEKA NEWMAN OKEREAFOR', ha='center', fontsize=10, color='gray')

plt.tight_layout()
plt.show()
```
